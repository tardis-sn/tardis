# TARDIS Quickstart: SDec and LIV Plots

This notebook runs a basic TARDIS simulation and generates Spectral Decomposition (SDec) and Line Interaction Volume (LIV) plots to analyze the output spectra.

```python
import tardis
from tardis import run_tardis
import matplotlib.pyplot as plt
# %matplotlib inline\n
print(f"TARDIS Version: {tardis.__version__}")
```

```python
# 1. Run the Quickstart Simulation
sim = run_tardis('quickstart_data/quickstart.yml')
```

## Spectral Decomposition (SDec) Plot
The SDec plot shows the contribution of different elements or ions to the final output spectrum.

```python
# 2. Generate SDec Plot
from tardis.visualization import SDECPlotter

sdec_plotter = SDECPlotter.from_simulation(sim)
fig_sdec = sdec_plotter.generate_plot()
plt.show()
```

## Line Interaction Volume (LIV) Plot
The LIV plot visualizes where specific line interactions occurred within the simulated supernova ejecta.

```python
# 3. Generate LIV Plot
from tardis.visualization import LIVPlotter

liv_plotter = LIVPlotter.from_simulation(sim)
fig_liv = liv_plotter.generate_plot()
plt.show()
```
