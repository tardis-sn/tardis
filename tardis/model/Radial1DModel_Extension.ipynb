```python
def to_csvy(fname, model, trad=False, dilution=False):
    with open(fname, 'w') as f:
        f.write('---\n')
        f.write('name: new_csvy.csvy\n')
        f.write('model_density_time_0: %f s\n'%(model.time_explosion.value))
        f.write('model_isotope_time_0: 0.00 s\n')
        f.write('description: Radial1DModel to csvy\n')
        f.write('tardis_model_config_version: v1.0\n')
        f.write('datatype:\n')
        f.write('  fields:\n')
        f.write('    -  name: velocity\n')
        f.write('       unit: km/s\n')
        f.write('       desc: velocities of shell outer bounderies.\n')
        f.write('    -  name: density\n')
        f.write('       unit: g/cm^3\n')
        f.write('       desc: density within shell with corresponding outer velocity.\n')
        if trad:
            f.write('    -  name: t_rad\n')
            f.write('       unit: K\n')
            f.write('       desc: radiative temperature within shell with corresponding outer velocity.\n')
        if dilution:
            f.write('    -  name: dilution_factor\n')
            f.write('       desc: dilution factor within shell with corresponding outer velocity.\n')
        abund_names = [nucname.name(Z) for Z in model.abundance.index]
        for name in abund_names:
            f.write('    -  name: %s\n'%(name))
            f.write('       desc: fractional %s abundance.\n'%(name))

        f.write('v_inner_boundary: %f km/s\n'%(model.v_boundary_inner.to('km/s').value))
        f.write('v_outer_boundary: %f km/s\n'%(model.v_boundary_outer.to('km/s').value))
        f.write('---\n')
        
        fieldnames = ['velocity', 'density']
        if trad: fieldnames.append('t_rad')
        if dilution: fieldnames.append('dilution_factor')
        for name in abund_names:
            fieldnames.append(name)
        for i in range(len(fieldnames)):
            if i == len(fieldnames) - 1:
                f.write('%s\n'%(fieldnames[i]))
            else:
                f.write('%s,'%(fieldnames[i]))
        
        for i in range(len(fieldnames)):
            if i == len(fieldnames) - 1:
                f.write('0.0\n')
            else:
                f.write('0.0,')
        
        csv = model.abundance.T
        csv.insert(0,'velocity',model.v_outer.to('km/s'))
        csv.insert(1,'density',model.density)
        if trad: csv.insert(2,'t_rad',model.t_rad)
        if dilution: csv.insert(3,'dilution_factor',model.dilution_factor)
        csv.to_csv(path_or_buf=f, index=False, index_label=False, header=False)
        
        f.close()
    return


def inject_element(model, element, mass):
    new_model = model
        
    abund_names = [nucname.name(Z) for Z in new_model.abundance.index]
    element_Z = nucname.name_zz[element]
    
    #check if injection element is missing from csvy
    #and add a row if it is missing.
    if not element in abund_names:
        new_abund_row = pd.Series(np.zeros(len(new_model.abundance.iloc[0])), name=element_Z)
        new_model.abundance = new_model.abundance.append(new_abund_row)
        new_model.abundance = new_model.abundance.sort_index()
        
    shell_masses = model.volume * model.density
    shell_masses = shell_masses.to('Msun')

    injected_mass = 0*u.Msun
    injection_index = new_model.abundance.shape[1] - 1
    
    #check if converting the next shell to desired element
    #will not yet reach the desired injection mass. If so,
    #proceed to convert next shell to desired element.
    #Loop does not run when next shell would overshoot
    #desired injection mass. This is a special case where
    #a fraction of the next shell must be converted and 
    #the velocity structure modified.
    while(injected_mass.value + shell_masses[injection_index].value < mass.value):
        # modify the abundance dataframe
        print(injection_index)
        col_length = len(new_model.abundance.iloc[:,injection_index])
        new_model.abundance.iloc[:,injection_index] = np.zeros(col_length)
        new_model.abundance.iloc[:,injection_index].loc[element_Z] = 1.0
        
        #update the injected mass
        injected_mass = injected_mass + shell_masses[injection_index]
        
        #update the injection index
        injection_index = injection_index - 1
    
    #handle the edge case of the last shell where
    #only a fraction of the shell is converted to
    #injection element.
    density = new_model.density[injection_index]
    r_outer = new_model.r_outer[injection_index]
    mass_diff = mass - injected_mass
    
    #Extend the adjacent outer shell inwards
    #by decreasing the next v_outer.
    new_r_outer = (r_outer**3 - 3 / (4 * np.pi * density) * mass_diff.to('g'))**(1.0/3.0)
    new_r_outer = new_r_outer.to('km')
    new_v_outer = new_r_outer / new_model.time_explosion
    
    print(new_v_outer)
    
    #+1 to the index because of inner boundary.
    new_model.velocity[injection_index + 1] = new_v_outer
    
    return new_model
    
    
    
```
