# Demo for Saving Plots as PNG Script

Running the generate_plots.py script for various different configuration files from tardis-setups repository. The script will save the plots as PNG files and those files are displayed in this notebook.

```python
from tardis.io.atom_data import download_atom_data
download_atom_data('kurucz_cd23_chianti_H_He_latest')
```

```python
import os
import sys
from IPython.display import Image, display
```

## 2022/Williamson_19ewu

```python
config = "2022/19ewu.yml"
atom_data = "kurucz_cd23_chianti_H_He_latest.h5"
output_path = "2022"
```

```python
os.system(f"python generate_plots.py --config {config} --atom_data {atom_data} --output_path {output_path}")
```

```python
display(Image(filename="2022/SDEC_Real.png"))
```

```python
display(Image(filename="2022/LIV_Real.png"))
```
