```python
# Getting Data

import urllib.request
from bs4 import BeautifulSoup
 
# here we have to pass url and path
# (where you want to save your text file)
urllib.request.urlretrieve("https://www.nndc.bnl.gov/nudat3/getdecaydataset.jsp?nucleus=56FE&dsid=56co%20ec%20decay","html_val.txt")
 
file = open("html_val.txt", "r")
contents = file.read()
soup = BeautifulSoup(contents, 'html.parser')
 
f = open("temp.txt", "w")
 
# traverse paragraphs from soup
for data in soup.find_all("pre"):
    lines = data.get_text()
    f.writelines(lines)
 
f.close()

# Removing blank lines 
# opening and creating new .txt file
with open("temp.txt", 'r') as r, open('Co56_ensdf.txt', 'w') as o:
    for line in r:
        #strip() function
        if line.strip():
            o.write(line)
f = open("Co56_ensdf.txt", "r")
print("New text file:\n",f.read())
```

```python
import requests
import json

url = "https://www.nndc.bnl.gov/radlist/radlist2.jsp"
# used https://www.nndc.bnl.gov/nudat3/indx_dec.jsp to get this file in previous block  or just upload and run the code don't run the above block if you have ENSDF file already
filename = "./Co56_ensdf.txt"  

with open(filename, "rb") as f:
    data = {'mydata': f.read(),'format':'json','submit':'Submit'}
    response = requests.post(url, data=data)
try:
    print("POST request successful")
    response_data = json.loads(response.text)
    print(response_data)
except json.decoder.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")
    print("Raw response content:")
   
with open('file.json', 'w') as f:
        json.dump(response_data, f)
print("Response data saved to file.json")

# delete temporary files 
import os 
os.remove("html_val.txt")
os.remove("temp.txt")
```
