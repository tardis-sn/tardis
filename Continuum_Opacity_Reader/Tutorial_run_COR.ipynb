Welcome to the Continuum Opacity Source Reader. This Notebook is a step by step walkthrough of how we can use this module to read scientific literature data from
.gz files and returns a Pandas DataFrame that contains the required data to be fed into the Carsus engine or be directly used.

The Continuum Opacity Source Reader is a file meant to read data from Scientific literature in the form of .gzip files. The file has to be imported from the Continuum Opacity source Reader Module. The module contains a single class : OpacityReader which in turn contains 2 functions :
1. ##__init__(file_path(required, takes a string argument), masking_value(optional, default = None, takes integer value), skip_rows(optional, default = 0, takes integer value)##: This is the constructor file of the program and is used to intialise the class variables that the program shall later use.
2. ##read()##: This is the function that actually reads the data from the file and returns a Pandas Dataframe of the data collected from the file. The file currently only reads from .gz files.|

Now that we are introduced to the project as well as it's methods, let us try to use the Continuum Opacity Source Reader to read fron a source file, that will be s92.201.gz file that is stored in the Datasets folder.

First we shall import the Continuum Opacity Source Reader.

```python
from Continuum_Opacity_Reader_test import OpacityReader
```

Now that we have the module imported, let us try to use it to read the s92.201.gz fiole. For that we shall first get the location of the required file and shall use the OS module for that.

```python
import os

file_path = os.path.join("..","Dataset","s92.201.gz")

#Let's print to see if it's working:
print(f"The file Path is {file_path}")
```

Now that we have the daqtaset path loaded, let us try to read data from it using the Continuum Opacity Source Reader. We shall the program by creating the object with the file_path with the file_path we found using os.path.join() method and shall use the masking_value = 20 and skip_rows = 120

```python
reader_new = OpacityReader(file_path,masking_value=120, skip_rows=20)
#Next we shall call the read function to read from the file_path and it will return a dataset to us containing the data that we needed.
df = reader_new.read()
#let us print the dataset next.
df.head()
```

```python
# Now let us try to check thye dataset and dataquality as a whole:
print("--- Data Integrity Check ---")

# 1. Check for Nulls (Legacy data often has gaps)
null_count = df.isnull().sum().sum()
print(f"Total Null values: {null_count}")

# 2. Verify Data Types (Ensuring everything is numeric)
print("\nColumn Data Types:")
print(df.dtypes)

# 3. Quick Stats (Ensuring log values are in a physical range)
print("\nPhysical Range Summary:")
display(df.describe().loc[['min', 'max']])
```
