# Generate SDEC and LIV Plots from TARDIS Configurations

This notebook demonstrates how to use the `plot_from_config.py` script to generate SDEC (Spectral Diagnostic Emission Correlation) and LIV (Line Interaction Visualization) plots from various TARDIS simulation configurations.

## Setup

First, let's import necessary libraries and set up the environment.

```python
import os
from IPython.display import display, Image
```

```python
script_path = 'plot_from_config.py'
output_path = '../tardis-demo/out/'
```

## Configuration 1: 19ewu.yml (Williamson 19ewu)

Generate plots for the 19ewu supernova configuration.

```python
config = '../tardis-setups/2022/williamson_19ewu/19ewu.yml'

os.system(f"python {script_path} {config} --output-dir {output_path}")
```

```python
display(Image(filename=f"{output_path}/19ewu_sdecplotter_real.png"))
display(Image(filename=f"{output_path}/19ewu_sdecplotter_virtual.png"))
display(Image(filename=f"{output_path}/19ewu_livplotter_real.png"))
display(Image(filename=f"{output_path}/19ewu_livplotter_virtual.png"))
```

## Configuration 2: tardis_example.yml (2014_kerzendorf_sim/appendix_A1)

```python
config = '../tardis-setups/2014/2014_kerzendorf_sim/appendix_A1/tardis_example.yml'

os.system(f"python {script_path} {config} --output-dir {output_path}")
```

```python
display(Image(filename=f"{output_path}/tardis_example_sdecplotter_real.png"))
display(Image(filename=f"{output_path}/tardis_example_sdecplotter_virtual.png"))
display(Image(filename=f"{output_path}/tardis_example_livplotter_real.png"))
display(Image(filename=f"{output_path}/tardis_example_livplotter_virtual.png"))
```

## Summary

We have successfully generated and displayed SDEC and LIV plots for two different TARDIS simulation configurations:
1. **19ewu.yml** – Williamson 19ewu supernova model
2. **tardis_example.yml** – example configuration

The `plot_from_config.py` script provides a convenient way to:
- Run TARDIS simulations with given configurations
- Generate SDEC and LIV visualization plots
- Save plots in PNG or PDF format
- Use the configuration filename as the default output prefix

This makes it easy to explore and compare simulation results across the tardis-setups repository.
