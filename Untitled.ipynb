```python
from tardis import run_tardis
from tardis.io.atom_data.util import download_atom_data
```

```python
download_atom_data('kurucz_cd23_chianti_H_He')
```

```python
from tardis.visualization import (
    shell_info_from_simulation,
    shell_info_from_hdf,
    LineInfoWidget,
)
```

```python
simul = run_tardis('30d.yml')
```

```python
%pylab inline

spectrum = simul.runner.spectrum
spectrum_virtual = simul.runner.spectrum_virtual
spectrum_integrated = simul.runner.spectrum_integrated

figure(figsize=(10,6))
plot(spectrum.wavelength, spectrum.luminosity_density_lambda, label='normal packets')
plot(spectrum.wavelength, spectrum_virtual.luminosity_density_lambda, label='virtual packets')
plot(spectrum.wavelength, spectrum_integrated.luminosity_density_lambda, label='formal integral')
xlabel('Wavelength [$\AA$]')
ylabel('Luminosity [erg/s/$\AA$]')
legend()
xlim(3000, 9000)
```

```python
line_info_widget = LineInfoWidget.from_simulation(simul)
line_info_widget.display()
```
