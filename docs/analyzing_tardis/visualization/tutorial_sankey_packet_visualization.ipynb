# Sankey Packet Visualization Tutorial

This tutorial demonstrates how to visualize Monte Carlo packet interactions in TARDIS using Sankey diagrams. We'll explore packet histories, create interactive visualizations, and interpret the physics behind the flows.

## Introduction to Packet Tracking

In TARDIS, Monte Carlo packets represent photons propagating through the supernova ejecta. Each packet undergoes various interactions:

- **EMISSION**: Packet created/emitted
- **LINE**: Absorption/emission at spectral lines
- **ESCATTERING**: Electron scattering
- **CONTINUUM_PROCESS**: Continuum processes
- **BOUNDARY**: Boundary interactions (escape/enter shells)

To track packet histories, enable `config.montecarlo.tracking.track_rpacket = True` in your simulation configuration.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tardis
from tardis import run_tardis
from tardis.visualization import (
    plot_packet_history_sankey,
    plot_multi_packet_sankey_after_carbon,
    create_interactive_sankey_panel
)

# Enable interactive plots in notebook
%matplotlib inline
```

## Running a Simulation with Packet Tracking

First, let's run a TARDIS simulation with packet tracking enabled. We'll use a simple configuration that includes carbon lines for the physics insight example.

```python
# Load configuration and enable tracking
config = tardis.io.config_reader.Configuration.from_yaml('tardis_example.yml')

# Enable packet tracking
config.montecarlo.tracking.track_rpacket = True

# Reduce packets for faster demo (optional)
config.montecarlo.no_of_packets = 1e4
config.montecarlo.last_no_of_packets = 1e4

# Run simulation
sim = run_tardis(config, show_progress_bars=False)

print(f"Simulation completed with {len(sim.transport.transport_state.tracker_full_df)} tracked events")
```

## Single Packet Sankey Diagram

Let's visualize the interaction history of a single packet. The Sankey diagram shows the flow of packet states from emission to escape.

```python
# Create Sankey diagram for a single packet
fig = plot_packet_history_sankey(sim)
fig.show()

# The diagram shows:
# - START: Packet emission
# - Interaction types: LINE, ESCATTERING, etc.
# - END: Packet escape
# Flow thickness represents the sequence of events
```

## Interpreting Sankey Diagrams

**Key Insights from Packet Flows:**

1. **Dominant Processes**: Thick flows indicate common interaction types
2. **Packet Fate**: Most packets escape, but some are absorbed
3. **Line Interactions**: Show spectral line absorption/emission
4. **Scattering**: Electron scattering redistributes energy
5. **Boundary Effects**: Shell transitions (can be filtered out)

**Physics Interpretation:**
- LINE interactions represent radiative transfer through spectral lines
- ESCATTERING thermalizes the radiation field
- CONTINUUM processes handle bound-free/free-free transitions
- The diagram reveals the packet's energy deposition history

## Multi-Packet Carbon Interaction Flows

This advanced visualization aggregates all packets that interact with Carbon I lines, showing the collective behavior after carbon absorption.

```python
# Create aggregated Sankey for packets after Carbon I absorption
try:
    fig_carbon = plot_multi_packet_sankey_after_carbon(sim)
    fig_carbon.show()
    
    print("This diagram shows the flow of packets after their first Carbon I line interaction.")
    print("Flow thickness indicates how many packets follow each path.")
except ValueError as e:
    print(f"Carbon analysis not available: {e}")
```

## Physics Insights from Carbon Flows

**Carbon I Interactions in Supernovae:**

- Carbon lines are important coolants in Type Ia supernovae
- C I absorption removes photons from the blue part of the spectrum
- Post-absorption behavior shows energy redistribution
- Thick flows to ESCAPE indicate efficient cooling
- Flows to LINE show continued line interactions
- This reveals the role of carbon in shaping the spectrum

## Performance Considerations

For large simulations with many packets:
- Packet tracking increases memory usage
- Limit analysis to subsets of packets
- Use boundary filtering to reduce noise
- Consider sampling packets for large datasets
