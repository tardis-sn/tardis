# TARDIS Quickstart: SDec and LIV Plots

This notebook demonstrates how to run a basic TARDIS simulation and produce two important diagnostic visualizations:

- **SDec (Spectral Decomposition)** — shows which elements and ions contribute to the synthetic spectrum.
- **LIV (Last Interaction Volume)** — shows where in the ejecta the last photon–packet interaction occurred.

Both plots are generated using the `SDECPlotter` and `LIVPlotter` classes from `tardis.visualization`.

```python
%matplotlib inline

import tardis
from tardis import run_tardis
from tardis.visualization import SDECPlotter, LIVPlotter

print(f"TARDIS version: {tardis.__version__}")
```

## Run the simulation

We use the standard `tardis_example.yml` configuration that ships with this notebook.
Replace the path with your own YAML file to visualize a different model.

```python
sim = run_tardis("tardis_example.yml")
```

## Spectral Decomposition (SDec) Plot

The SDec plot decomposes the synthetic spectrum by element or ion, showing which species
dominate the emission and absorption features.  We use `generate_plot_mpl` for a static
Matplotlib figure suitable for publication.

```python
sdec_plotter = SDECPlotter.from_simulation(sim)
ax = sdec_plotter.generate_plot_mpl(packets_mode="real")
ax.figure.tight_layout()
```

## Line Interaction Volume (LIV) Plot

The LIV plot shows the spatial origin of the last line interaction for each photon packet,
giving insight into the photospheric structure of the simulated supernova.

```python
liv_plotter = LIVPlotter.from_simulation(sim)
fig = liv_plotter.generate_plot()
fig.show()
```
