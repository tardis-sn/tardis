#### Customizing the output
One can override the default layout by passing `plasma_plot_config` and the `luminosity_plot_config` dictionaries in the run_tardis function. These dictionaries change the layout of the respective plots. The `colorscale` parameter is used to change the colorscale used in the plasma plots. One can also not show the plots by passing `show_plot=False` in the `run_tardis` function. 

```python
from tardis import run_tardis
sim = run_tardis('tardis_example.yml', plasma_plot_config = {"height": 500}, luminosity_plot_config = {"height": 500})
```
